### Farmer Brown and Farmer Green each plant 100 acres of corn under similar conditions. Farmer Brown's yield was 84 bushels per acre with a standard deviation of 5 bushels. Farmer Green's yield was 80 bushels per acre with a standard deviation of 6 bushels. What is the 90% confidence interval for the mean difference in yield between the two farms?

### <strong><U>SOLUTION:</strong></U> 

```python
### Farmer Brown and Farmer Green each plant size is 100 i.e.,
### population means with two samples is more than 30. Then he central limit theorem can be applied here.
### Both the Farmer's Brown and Green respectively are as follows:


import scipy.stats
import math

### Farmer Brown Sample 1
br_ac = 100 

br_bush= 84

br_sd = 5 

### Farmer Green Sample 2   
gr_ac = 100 

gr_bush= 80 

gr_sd= 6

Level_of_Significance = 0.1

Degrees_of_freedom= br_ac+gr_ac-2 # 198

Critical_value = scipy.stats.t.ppf(q=1-.10/2,df=198) # 1.652586



Pooled_variance_Sp2 = (((br_ac-1)*(br_sd**2))+((gr_ac-1)*(gr_sd**2)))/(br_ac+gr_ac-2) 

Pooled_standard_deviation_sp = math.sqrt(Pooled_variance_Sp2)#5.5227

E =  Critical_value*Pooled_standard_deviation_sp*math.sqrt((1/br_ac)+(1/gr_ac))

#90% confidence level is = (x1 -x2) ± E

Lower_limit = (br_bush -gr_bush) - E 

Upper_limit = (br_bush -gr_bush) + E 

print(f'90% Confidence level is = {Lower_limit} < µ1-µ2 < {Upper_limit}')
```

### <strong><U>PROBLEM 3.</strong></U> 

### SJSU bought 36 bulbs for its classroom projectors. The manufacturer claimed that the bulbs should each last more than 800 hours. After using the bulbs, the university discovered that their mean lifetime was 816 hours with a standard deviation of 70 hours. At the 5% level of significance, should the university accept the manufacturer's claim?

### <strong><U>SOLUTION:</strong></U> 

Here the null hypothesis H0: µ = 800 hrs  

The alternative hypothesis Ha: µ > 800 hours.

### Given information of bulbs for its classroom projectors bought by SJSU

µ > 800

### Given

#### Mean lifetime of bulbs are:- x = 816

#### Standard Deviation is:- s = 70

n = 36

#### Test hypothesis are
H0 :   μ = 800
Ha :   μ > 800

Test statistic t = (x - μ)/(s/sqrt(n))
                 = (816-800)/(70/sqrt(36))
                 = 1.371428571 = 1.37

p value is , 
P(t >t0) = 0.0897            # T.DIST.RT(t0,df) (Using excel)
Since p > α                  # hence fail to reject H0

####  At the 5% level of significance , the university should not accept the manufacturer's claim because we have failed.

```python
import scipy.stats as st

#Null hypothesis: H0: µ = 800   
#Alternative hypotheses: Ha: µ > 800

mu = 800   # Population mean

n = 36   # Sample size

x_bar = 816 # Sample mean

sd = 70  # Sample standard deviation

los = 0.05 # level of significance

# Calculating the test statistic('t')

t = (x_bar - mu) / (sd / math.sqrt(n))

#Finding probability value 'p'

p = 1-st.t.cdf(t,n-1)

print("The t statistics : {0:.3f}".format(t))
print("The probability 'p' value : {0:.2f}".format(p))
print()
print("Since p is > level of significance, SJSU should accept the manufacturer's claim")
```

### A widget making machine is supposed to make widgets that are 0.05 inches thick. Ten widgets are tested each day to ensure the machine is properly calibrated. Today's test averaged 0.053 inches thick with a standard deviation of 0.003. At the 5% level of significance, is the machine properly calibrated?

### <strong><U>SOLUTION:</strong></U> 

### A Widget making machine is supposed to make widgets there are 0.05 inches thick

#### Sample Size: n = 10

#### Sample Mean: x = 0.053

#### Sample Standard Deviation: 0.003

#### Significance Level: σ = 5% = 0.05

#### Null Hypothesis: Ho = µ = 0.05 (where machine is calibrated properly)

#### Alternative Hypothesis: Ha = µ is not equal to 0.05 (where machine is not calibrated properly)

#### t = x-µ/(s/sqrt(n)) = 0.053 - 0.05/(0.003/sqrt(10)) = 3.162

#### Critical Value of 't' at 0.05 level of significance for 9 , degree of freedom for two tailed test

#### t (0.005 , 9) = 2.262

#### Calculated |t| = 3.162

#### Critical t = 2.262

#### Since Calculated |t| > Critical t



```python
#H0: µ = 0.05，the machine is properly calibrated   
#Ha: µ ≠ 0.05，the machine is not properly calibrated

thickness_mu = 0.05 #thickness
x_bar = 0.053 #Sample Mean
sd = 0.003  #  Sample standard deviation
n = 10  #Sample size

los = 0.05 # level of significance

# Calculating the test statistic('t')
t = (x_bar - thickness_mu) / (sd / math.sqrt(n))

#Finding probability value 'p'
p = 1-st.t.cdf(t,n-1)

print("The t statistics : {0:.3f}".format(t))
print("The probability 'p' value: {0:.2f}".format(p))
print()
print("At 5% level of significance, the machine is not properly calculated since p < level of significance.")
```

#### <I> <U>Conclusion<U> <I> 
#### Where we neglate the null Hypothesis (Ho) at 5% level of significance and conclude that machine is not properly calibrated because (µ is not equal to 0.05)

### Farmer Brown's cows have averaged 380 pounds in weight. He selects 50 of his cows to put on a new diet to make them gain weight. After being on the new diet, the selected cows weigh an average of 390 pounds with a standard deviation of 35.2 pounds. What are the null and alternative hypotheses? What is the lowest level of significance at which we can reject the null hypothesis?

### <strong><U>SOLUTION:</strong></U> 

### In the question, the information given for the Farmer Browns's cows:

#### Weight of the cows in pounds:- µ = 380  

#### Average weight of the cows, T = 390

#### Standard Deviation of the wieght:- s = 35.2

#### Cows selected:- n = 50

#### The null and alternative hypothesis is,

H0 : µ = 390
Ha : µ is not equal to 390

#### This is the two tailed test .  

#### Lets take Test statistic = t

t = (x - µ ) / (s / sqrt(n))

t = (390 - 380) / (35.2 / sqrt(50))

#### Test statistic (t) = 2.01

df = n - 1 = 50 - 1 = 49

P value = 0.05

alpha = 0.10

0.05 < 0.10

P value < alpha



```python
print("Null hypothesis, H0: mu = 380")
print("Alternate hypothesis, Ha: mu > 380")
```

```python
mu = 380    # Population Mean
n = 50      # Sample Size
x_bar = 390 # Sample Mean
sd = 35.2   # Sample Standard Deviation

# Finding the test statistic('z')
z = (x_bar - mu) / (sd/math.sqrt(n))

#Finding probability value 'p'
p = 1 - st.norm.cdf(z)
print("The probability 'p' value: {0:.4f}".format(p))
print()
print("The lowest level of significance at which we can reject the null hypothesis is {0:.3f} or {1:.1f}%".format(p, p*100))
```

### Using the table from Problem 8, plot the 3- and 5-year moving averages.

### <strong><U>SOLUTION:</strong></U> 

```python
def moving_average(years):
    i=0
    moving_averages_ = []

    while i < len(snowfall) - years + 1:
        window = snowfall[i : i + years]
        # Calculate the average of current window
        window_average = round(sum(window) / years, 2)
        # Store the average of current
        # window in moving average list
        moving_averages_.append(window_average)
        # Shift window to right by one position
        i += 1    
    return moving_averages_

```

```python
y_3 = moving_average(3)
print(y_3)
```

```python
y_5 = moving_average(5)
print(y_5)
```

```python
plt.plot(year,snowfall,'o-',label='snowfall',color='darkcyan')
plt.plot(year[len(year)-len(y_3):],y_3,'o-',label='3 years moving avg',color='maroon')
plt.plot(year[len(year)-len(y_5):],y_5,'o-',label='5 years moving avg')
plt.title('Plot for 3 and 5 years moving averages')
plt.xlabel('Years')
plt.ylabel('snowfall')
plt.legend()
plt.rcParams["figure.figsize"] = (15, 10)
plt.show()
```

### <strong><U>PROBLEM 10.</strong></U> 

###  Using the table from Problem 8, exponentially smooth the data using alpha values 0.3 and 0.1

### <strong><U>SOLUTION:</strong></U> 

```python
def expo_smoothing(x_values,y_values,alpha):
    smooth_l = []
    for i in range(len(x_values)):
        y = y_values[i]
        if i == 0:
            smooth = y
        else:
            smooth = alpha*y + (1 - alpha)*smooth
        smooth_l.append(smooth)
    return smooth_l
```

```python
es_3 = expo_smoothing(year, snowfall, alpha=0.3)
```

```python
print(es_3)
```

```python
es_1 = expo_smoothing(year, snowfall, alpha=0.1)
print(es_1)
```
